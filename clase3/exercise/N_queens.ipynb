## Generaciรณn de vecinos

En los algoritmos de bรบsqueda local, la idea es moverse en el espacio de soluciones explorando vecinos del estado actual.

**ยฟQuรฉ es un vecino en el problema de las N reinas?**

Dado un estado (una configuraciรณn particular de reinas en el tablero), un vecino es otro estado que resulta de hacer una pequeรฑa modificaciรณn en el estado actual.

En nuestro caso, donde cada reina estรก en una columna, una forma natural de definir vecinos es:

1. Elegir una columna.
2. Mover la reina de esa columna a una fila diferente dentro de la misma columna.

**Ejemplo**

Si el estado actual es:

```python
board = [0, 4, 7, 5, 2, 6, 1, 3]
```

Un vecino puede ser:

`Mover la reina en la columna 2 (que estรก en fila 7) a la fila 0`, dando:

```python
neig_board = [0, 4, 0, 5, 2, 6, 1, 3]
```

A continuaciรณn, definamos una funciรณn que genera todos los vecinos posibles a partir de un estado dado:

```python
def generate_neighbors(state):
    """
    Generates all neighboring states from the current state.
    A neighbor is a state where one queen is moved to a different
    row within its column.
    
    Args:
        state (list): List representing the queen positions.
        
    Returns:
        list: List of neighboring states.
    """
    n = len(state)
    neighbors = []
    
    for col in range(n):
        current_row = state[col]
        for new_row in range(n):
            if new_row != current_row:
                neighbor = state.copy()
                neighbor[col] = new_row
                neighbors.append(neighbor)
                
    return neighbors

# Ejemplo de uso:
estado_actual = [0, 4, 7, 5, 2, 6, 1, 3]
vecinos = generate_neighbors(estado_actual)
print(f"Cantidad de vecinos generados: {len(vecinos)}")
print("Algunos vecinos:")
for vecino in vecinos[:5]:
    print(vecino)
```

**ยฟQuรฉ hace esta funciรณn?**

1. Para cada columna, prueba mover la reina a todas las filas distintas de la actual.
2. Copia el estado actual y modifica la posiciรณn para crear un vecino nuevo.
3. Devuelve una lista con todos los vecinos posibles.

## Funciรณn de costo

En los algoritmos de bรบsqueda local, cada estado (o soluciรณn parcial) tiene un *costo* que indica quรฉ tan lejos estรก de ser una soluciรณn vรกlida.

En el caso del problema de las N reinas, una soluciรณn vรกlida es aquella donde ninguna reina puede atacar a otra. Por lo tanto, una forma natural de definir la funciรณn de costo es:

> Sea:
> - N el tamaรฑo del tablero (y cantidad de reinas)
> - $s = [s_0, s_1, ..., s_{N-1}]$ un estado, donde $s_i \in \{ 0, 1, ..., N-1 \}$ representa la fila en la quie estรก la reina de la columna *i*.
>
> Entonces, la funciรณn de costo $C(s)$ mide el nรบmero total de pares de reinas que se atacan entre sรญ:
>
> $$C(s) = \sum_{i=0}^{N-1} \sum_{j=i+1}^{N-1} \delta(s_i, s_j)$$
> donde la funciรณn $\delta(s_i, s_j)$ se define como:
> $$ \delta(s_i, s_j) = \begin{cases} 1 & \text{si } s_i = s_j \quad \text{(misma fila)} \\ 1 & \text{si } |s_i - s_j| = |i - j| \quad \text{(misma diagonal)} \\ 0 & \text{en otro caso}\end{cases}  $$

Queremos minimizar esta funciรณn:
- Si el costo es `0`, ยกtenemos una soluciรณn vรกlida!
- Si el costo es mayor que `0`, hay conflictos que resolver.

ยฟCuรกndo dos reinas se atacan? Dos reinas se atacan si estรกn:

- En la misma fila
- En la misma diagonal (ascendente o descendente)
- โ ๏ธ No necesitamos preocuparnos por columnas, porque la representaciรณn del problema (una reina por columna) ya garantiza que no haya mรกs de una reina por columna.

La implementaciรณn en Python es:

```python
def cost(state):
    """
    Calculates the number of pairs of queens that are attacking each other.
    
    Args:
        state (list): A list where state[i] is the row of the queen in column i.
        
    Returns:
        int: Total number of conflicting pairs.
    """
    n = len(state)
    conflicts = 0
    for i in range(n):
        for j in range(i + 1, n):
            same_row = state[i] == state[j]
            same_diag = abs(state[i] - state[j]) == abs(i - j)
            if same_row or same_diag:
                conflicts += 1
    return conflicts

# Ejemplo de uso
state = [0, 4, 7, 5, 2, 6, 1, 3]  # Soluciรณn valida para N=8
print("Estado:", state) 
print("Costo:", cost(state))     

state = [0, 4, 7, 5, 2, 6, 1, 0]
print("Estado:", state)
print("Costo:", cost(state))
```

Esta funciรณn se usa para:
- Evaluar quรฉ tan buena es una soluciรณn actual.
- Comparar varios vecinos y elegir el de menor costo.
- Saber si ya llegamos a una soluciรณn (cuando el costo es 0).

## Visualizaciรณn del tablero

Una forma muy รบtil de comprender y depurar soluciones al problema de las N reinas es poder visualizar el estado del tablero. Para esto, podemos imprimirlo en consola usando caracteres ASCII, representando las casillas vacรญas y las reinas.

ยฟCรณmo lo representamos? Usamos los siguientes sรญmbolos:

- `โ` para indicar una reina
- `|` y `---` para simular los bordes del tablero

Definamos la funciรณn que nos permita armar la representaciรณn:

```python
def print_board_ascii(state: list):
    """
    Prints an ASCII representation of the N-Queens board.
    
    Each index in the input list represents the column,
    and the value at that index represents the row where the queen is placed.

    Args:
        state (list): A list of integers where state[i] is the row of the queen in column i.

    Example:
        For board = [0, 4, 7, 5, 2, 6, 1, 3], prints a visual 8x8 board with queens.
    """
    n = len(state)
    horizontal_border = "+" + "----+" * n
    for row in range(n):
        print(horizontal_border)
        row_str = ""
        for col in range(n):
            if state[col] == row:
                row_str += "| โ "
            else:
                row_str += "|    "
        row_str += "|"
        print(row_str)
    print(horizontal_border)

# Ejemplo de uso
state = [0, 4, 7, 5, 2, 6, 1, 3]  # Soluciรณn vรกlida para N=8
print_board_ascii(state)
```

## Algoritmo de gradiente descendente (hill climbing)

Tal como vimos en los videos, este algoritmo parte de un estado inicial y se mueve siempre al mejor vecino, es decir, al estado vecino con menor cantidad de conflictos (menor costo), esperando llegar a una soluciรณn sin conflictos.

**Estructura general del algoritmo**
1. Generar un estado inicial aleatorio.
2. Evaluar su costo.
3. Generar todos sus vecinos.
4. Elegir el vecino con menor costo.
5. Si es mejor que el actual, avanzar a ese estado.
6. Si no hay mejora posible โ nos quedamos en un รณptimo local.

Veamos la implementaciรณn:

```python
N = 8 # Nรบmero de reinas
MAX_STEPS = 1000 # Nรบmero maximo de iteraciones

def grad_desc(n, max_steps=1000) -> tuple:
    """
    Hill Climbing algorithm for solving the N-Queens problem.
    
    Args:
        n (int): Size of the board (number of queens).
        max_steps (int): Maximum number of steps before giving up.
        
    Returns:
        tuple: (final_state, final_cost)
    """

    # 1. Inicializamos un estado al azar
    current = initialize_random_state(n)
    # Calculamos el costo
    current_cost = cost(current)

    print("Estado inicial:", current)
    print("Costo:", current_cost)
    print_board_ascii(current)
    print("------------")
    
    last_step = max_steps
    for step in range(max_steps):
        # 3. Generamos todos sus vecinos.
        neighbors = generate_neighbors(current)
        # Calculamos todos los costos de los vecinos
        neighbor_costs = [cost(neighbor) for neighbor in neighbors]
        # Obtenemos el minimo costo
        min_cost = min(neighbor_costs)

        # Si todos los vecinos tienen el mismo valor o mayor, terminamos.
        if min_cost >= current_cost:
            last_step = step
            break

        # Elegimos el vecino con menor costo.
        best_index = neighbor_costs.index(min_cost)
        current = neighbors[best_index]
        current_cost = min_cost
        

    if current_cost == 0:
        print("โ ยกSe encontrรณ una soluciรณn!")
    else:
        print("โ No se encontrรณ una soluciรณn :(")

    print("Iteraciones ejecutadas:", last_step)
    print("Estado final:", current)
    print("Costo final:", current_cost)
    print_board_ascii(current)

    return current, current_cost
```

### Simmulated Annealing

Llegamos a la parte que **vos debรฉs desarrollar**. Ahora te toca implementar el algoritmo de bรบsqueda Simulated Annealing.

Para eso, debรฉs completar la funciรณn que aparece mรกs abajo.

> โ ๏ธ No modifiques los argumentos ni el tipo de salida: la funciรณn debe recibir exactamente los parรกmetros indicados y devolver el resultado especificado.

```python
def should_search(delta,temperature):
    return random.random() < exp(-delta / temperature)

def get_neighbour(neighbours):
    neighbours_size = len(neighbours)
    random_neighbour_index = random.randrange(neighbours_size)
    return neighbours.pop(random_neighbour_index)

def simulated_annealing(n: int, initial_temp: float, cooling_rate: float, max_steps: int = 1000) -> tuple:
    """
    Simulated Annealing algorithm for solving the N-Queens problem.
    
    Args:
        n (int): Size of the board (number of queens).
        initial_temp (float): Starting temperature.
        cooling_rate (float): How much the temperature decreases per step.
        max_steps (int): Maximum number of iterations.
    
    Returns:
        tuple: (final_state, final_cost)
    """
    current = initialize_random_state(n)
    current_cost = cost(current)
    temperature = initial_temp
    best_state = current
    best_cost = current_cost

    print("Estado inicial:", current)
    print("Costo:", current_cost)
    print_board_ascii(current)
    print("------------")

    last_step = 0
    for step in range(max_steps):
        last_step += 1
        current_cost = cost(current)
        neighbours = generate_neighbors(current)
        while len(neighbours) > 0:
            neighbour = get_neighbour(neighbours)
            neighbour_cost = cost(neighbour) 
            cost_diff = neighbour_cost - current_cost
            if cost_diff < 0:
                current = neighbour
                break
            else:
                if temperature > 0:
                    if should_search(cost_diff, temperature):
                        current = neighbour
                        break 
        
        if current_cost < best_cost:
            best_state = current
            best_cost = cost(current)
        
        if best_cost == 0:
            break
        
        temperature *= (1 - cooling_rate)

    if current_cost == 0:
        print("โ ยกSe encontrรณ una soluciรณn!")
    else:
        print("โ No se encontrรณ una soluciรณn :(")

    print("Iteraciones ejecutadas:", last_step)
    print("Estado final:", current)
    print("Costo final:", current_cost)
    print_board_ascii(current)

    return best_state, best_cost 
initial_temp = 1
cooling_rate = 0.05
```
